# Comandos do Pandas para melhorar a visualização de dados

Agradecimento a https://towardsdatascience.com/6-pandas-display-options-you-should-memories-84adf8887bc3

```
# Importação do Pandas
import pandas as pd
import numpy as np
```

```
# criando um dataframe para uso nos exemplos
df = pd.DataFrame(np.random.randn(500, 101))
df.info()
```

```
# apresentação com os padrões do Pandas
df
```

```
# mostra até 101 colunas sem corte com ... no centro
pd.options.display.max_columns = 101
df
```

```
# altera a quantidade de colunas a terem as informações serem mostradas 
pd.options.display.max_info_columns = 150
df.info()
```

```
# modifica a apresentação da precisão decimal
pd.options.display.precision = 2
df
```

```
# modifica a formatação dos campos
pd.options.display.float_format = '{:.1f}%'.format
df
```
